<h4>2.3 Filtering and reindexing</h4>

```python
import numpy as np
import pandas as pd

# Load dataset
df = pd.read_csv(r"C:\Users\nayif\python_nayef_omer2\week3\flavors_of_cacao.csv")
samples = df.shape[0]  # Number of rows (samples)
columns = df.shape[1]  # Number of columns

print("Number of samples (rows):", samples)
print("Number of columns:", columns)
print("\nDataset summary before processing:")
print(df.describe())  # Summarized view of the dataset

# Function to count NaN values in each column
def count_nan(dataframe):
    for col in dataframe.columns:
        missing_percentage = dataframe[col].isna().sum() / len(dataframe) * 100
        print(f"The column '{col}' is missing {missing_percentage:.2f}% of data!")

# Display NaN counts before dropping
print("\nMissing data before dropping NaN values:")
count_nan(df)

# Drop rows with NaN values and check again
df = df.dropna()

print("\nMissing data after dropping NaN values:")
count_nan(df)

# Function to normalize values to the range [0, 1]
def normalize_dataframe(dataframe):
    for column in dataframe.select_dtypes(include=[np.number]).columns:  # Only normalize numeric columns
        min_val = dataframe[column].min()
        max_val = dataframe[column].max()
        if max_val != min_val:  # Avoid division by zero
            dataframe[column] = (dataframe[column] - min_val) / (max_val - min_val)
        else:
            dataframe[column] = 0.5  # Assign a midpoint if min and max are the same

# Apply normalization
normalize_dataframe(df)
print("\nNormalized dataset summary:")
print(df.describe())
```
